```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
### Part A (10 pts)

# Create the input and output data for the network to train on, as discuessed above.
# Store the input data as a 4 column numpy array called x_data
# Store the output data as a 1 column numpy array called y_data

x_data = np.zeros((2**4, 4))
y_data = np.zeros((2**4, 1))
current_input = [0, 0, 0, 0]
current_row = 0
while current_row < x_data.shape[0]:
    # Store the current input in the current row
    x_data[current_row, :] = current_input
    y_data[current_row, :] = sum(current_input) % 2
    # Create the next input
    current_input[3] += 1
    for i in range(3, -1, -1):
        if current_input[i] == 2:
            current_input[i] = 0
            current_input[i - 1] += 1
    # Iterate to the next row
    current_row += 1
```

```python
### Part B (25 pts)

# 1 -- Network Initialization

# Initialize a simple sequential neural network mode
model = Sequential()

model.add(Dense(units=32, input_dim=4, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.summary()
```

```python
# 2 -- Network Compilation

# Compile. Loss function: binary crossentropy. Optimizer: stochastic gradient descent (SGD). Metrics: 'accuracy'
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
# Network training
model.fit(x_data, y_data, epochs=500, batch_size=3)
```

```python
# 4 -- Network Evaluation

### Save the final cost of the loss model in a varaible called 'final_cost'.
### Save the final cost of the model on the test set in a variable called 'test_accuracy'.
loss_and_metrics = model.evaluate(x_data, y_data, verbose=0)
loss_and_metrics
```

```python
final_cost = loss_and_metrics[0]
test_acc = loss_and_metrics[1]
print("Final test cost: ", final_cost)
print("Final test accuracy: ", test_acc)
```
